## Cleaning Data

### 1.Changing all aplhabets into lower case
  

```python
def data_preprocess(comments):
    preprocessed_comments = [i.lower() for i in comments]
    return preprocessed_comments

space = data_preprocess(Trump_data)
space
```

### 2. Removing spaces in beginning and end

```python
def data_spaces(comments):
    preprocessed_comments = [i.strip() for i in comments]
    return preprocessed_comments

Trump_data_space = data_spaces(space)
Trump_data_space
```

### 3.Removing Punctuations

```python
import string
def remove_punctuations(comments):
    
    translator = [i.translate(str.maketrans('','' ,string.punctuation)) for i in comments]
    return translator

Trump_data_punctuation = remove_punctuations(Trump_data_space)
Trump_data_punctuation
```

## Number of words in Trump Data

```python
def number_of_words(word_list):
    result_list=[]
    line_list =[line.split() for line in word_list]
    print(line_list)
    for j in line_list:
        result_list.extend(j) 
    return len(result_list)

Trump_data_count = number_of_words(Trump_data_punctuation)

```

```python
print("Number of words in Trump data is = ",Trump_data_count)
```

### Total positive, negative, stop and other count

```python
def create_word_list(word_list):
    result_list = []
    line_list =[ line.split() for line in word_list]
    for j in line_list:
        result_list.extend(j) 
    return result_list

Trump_data_word_list = create_word_list(Trump_data_punctuation)

def calculate_word_count(Trump_data_word_list):
    pos_word_count = 0
    neg_word_count = 0
    stop_word_count= 0
    other_word_count = 0
    for i in Trump_data_word_list:
        if i in positive:
            pos_word_count += 1
        elif i in negative:
            neg_word_count += 1
        elif i in stopwords:
            stop_word_count +=1
        else:
            other_word_count +=1
            
    return pos_word_count,neg_word_count,stop_word_count,other_word_count

Trump_pos_count, Trump_neg_count,Trump_stop_word_count,Trump_other_word_count = calculate_word_count(Trump_data_word_list)
print("Trump positive word count = ",Trump_pos_count)
print("Trump negative word count = ", Trump_neg_count)
print("Trump stop word count = ", Trump_stop_word_count)
print("Trump other word count = ",Trump_other_word_count)
```

### Ratio between positive, negative, stop and other word count to total

```python
def ratioFunction(Trump_pos_count,Trump_neg_count,Trump_stop_word_count,Trump_other_word_count,Trump_data_count):
   pos_ratio=(Trump_pos_count/Trump_data_count)
   print('Ratio of Positive word ',Trump_pos_count,'to Total data',Trump_data_count,'is',pos_ratio)
   neg_ratio=(Trump_neg_count/Trump_data_count)
   print('Ratio of Negative word ',Trump_neg_count,'to Total data',Trump_data_count,'is',neg_ratio)
   stop_ratio=(Trump_stop_word_count/Trump_data_count)
   print('Ratio of Stop word ',Trump_stop_word_count,'to Total data',Trump_data_count,'is',stop_ratio)
   other_ratio=(Trump_other_word_count/Trump_data_count)
   print('Ratio of other word ',Trump_other_word_count,'to Total data',Trump_data_count,'is',other_ratio)
   all_to_total=(Trump_pos_count+Trump_neg_count+Trump_stop_word_count+Trump_other_word_count/Trump_data_count)
   print('Ratio of all words to Total data is',all_to_total)
ratioFunction(Trump_pos_count,Trump_neg_count,Trump_stop_word_count,Trump_other_word_count,Trump_data_count)
```

### Ratio of positive VS Negative count

```python
def ratioFunction(Trump_pos_count,Trump_neg_count):
   pos_neg_ratio=(Trump_pos_count/Trump_neg_count)
   print('Ratio of Positive word ',Trump_pos_count,'to Negative word',Trump_neg_count,'is',pos_neg_ratio)
ratioFunction(Trump_pos_count,Trump_neg_count)
```
